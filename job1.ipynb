```python
population_columns = [
    '1970 Population', '1980 Population', '1990 Population',
    '2000 Population', '2010 Population', '2015 Population',
    '2020 Population', '2022 Population'
]

continent_population_stats = df.groupby('Continent')[population_columns].agg(
    ['mean', 'median', 'var', 'std']
)

print("Population Statistics by Continent:")
print(continent_population_stats)
```

```python

cumulative_growth_df = df[['Country/Territory', '1970 Population']].copy()
cumulative_growth_df
```

```python
for col in population_columns[1:]:
    cumulative_growth_df[f'{col} Cumulative Growth (%)'] = (
        (df[col] - df['1970 Population']) / df['1970 Population']
    ) * 100

print("Cumulative Population Growth for Each Country (relative to 1970 Population):")
print(cumulative_growth_df.head())
```

```python
new_columns = []
for col in df.columns:
    if ' Population' in col and col.replace(' Population', '').isdigit() and len(col.replace(' Population', '')) == 4:
        new_columns.append(col.replace(' Population', ''))
    else:
        new_columns.append(col)

df.columns = new_columns

print("DataFrame columns after renaming:")
print(df.head())
```

```python
available_years = [col for col in df.columns if col.isdigit() and len(col) == 4]

requested_years = ['1900', '2022']

for year in requested_years:
    if year not in available_years:
        print(f"Warning: Population data for year {year} is not available in the dataset. Skipping visualization for {year}.")

if '2022' in available_years:
    plt.figure(figsize=(15, 10))
    plt.suptitle('Distribution of 2022 Population by Continent', fontsize=16)

    continents = df['Continent'].unique()
    num_continents = len(continents)

    ncols = 2
    nrows = (num_continents + ncols - 1) // ncols

    for i, continent in enumerate(continents):
        plt.subplot(nrows, ncols, i + 1)
        continent_data = df[df['Continent'] == continent]['2022']
        sns.histplot(continent_data, kde=True, bins=10, color='skyblue')
        plt.title(f'{continent} (2022)')
        plt.xlabel('Population')
        plt.ylabel('Number of Countries')
        plt.ticklabel_format(style='plain', axis='x')

    plt.tight_layout(rect=[0, 0.03, 1, 0.96])

    plt.savefig('Visualize_2022_population_by_continent_using_histograms.png')
    plt.show()
```

```python
population_year_cols = available_years

df_melted = pd.melt(df, id_vars=['Continent'], value_vars=population_year_cols, var_name='Year', value_name='Population')

df_melted['Year'] = df_melted['Year'].astype(int)

print("Melted DataFrame (df_melted):")
print(df_melted.head())
```

```python
continent_population_trend = df_melted.groupby(['Continent', 'Year'])['Population'].sum().reset_index()

print("Continent Population Trend DataFrame (head):")
print(continent_population_trend.head())
```

```python
plt.figure(figsize=(12, 7))
sns.lineplot(data=continent_population_trend, x='Year', y='Population', hue='Continent', marker='o')
plt.title('Population Trends by Continent Over Time', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Population', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Continent')
plt.tight_layout()
plt.savefig('Population_Trends_by_Continent_Over_Time.png')
plt.show()
```
