```python
population_columns = [
    '1970 Population', '1980 Population', '1990 Population',
    '2000 Population', '2010 Population', '2015 Population',
    '2020 Population', '2022 Population'
]

continent_population_stats = df.groupby('Continent')[population_columns].agg(
    ['mean', 'median', 'var', 'std']
)

print("Population Statistics by Continent:")
print(continent_population_stats)
```

```python
df.shape
```

```python

cumulative_growth_df = df[['Country/Territory', '1970 Population']].copy()
cumulative_growth_df
```

```python
for col in population_columns[1:]:
    cumulative_growth_df[f'{col} Cumulative Growth (%)'] = (
        (df[col] - df['1970 Population']) / df['1970 Population']
    ) * 100

print("Cumulative Population Growth for Each Country (relative to 1970 Population):")
print(cumulative_growth_df.head())
```

```python
new_columns = []
for col in df.columns:
    if ' Population' in col and col.replace(' Population', '').isdigit() and len(col.replace(' Population', '')) == 4:
        new_columns.append(col.replace(' Population', ''))
    else:
        new_columns.append(col)

df.columns = new_columns

print("DataFrame columns after renaming:")
print(df.head())
```
